```python
# Figure 3: scattterplot comparing two stability related attributes
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

import os
os.getcwd()

DATA_DIR = Path("JBI100_Visualization_project_group_80/data/raw")
demographics = pd.read_csv(DATA_DIR / "demographics_data.csv")
economy = pd.read_csv(DATA_DIR / "economy_data.csv")

print(demographics.columns)
print(economy.columns)

gdp_col = "GDP per capita"
literacy_col = "Literacy rate, adult total"

df = pd.merge(demographics, economy, on="Country", how="inner")
df = df.dropna(subset=[gdp_col, literacy_col]).copy()

plt.figure(figsize=(7, 5))
plt.scatter(df[gdp_col], df[literacy_col])
plt.xlabel("GDP per capita")
plt.ylabel("Literacy Rate")
plt.title("GDP per capita vs Literacy Rate")
plt.grid(True, alpha=0.3)
plt.savefig("../report/stability_scatter.png", dpi=300, bbox_inches="tight")
plt.show()
```
