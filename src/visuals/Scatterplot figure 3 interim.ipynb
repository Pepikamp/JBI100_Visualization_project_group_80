```python
# Figure 3: scattterplot comparing two stability related attributes
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

DATA_DIR = Path("../../data/raw")
demographics = pd.read_csv(DATA_DIR / "demographics_data.csv")
economy = pd.read_csv(DATA_DIR / "economy_data.csv")

df = pd.merge(demographics, economy, on="Country", how="inner")
df_clean = df.dropna(subset=["Infant_Mortality_Rate", "Real_GDP_per_Capita_USD"])

plt.figure(figsize=(7, 5))
plt.scatter(df_clean["Real_GDP_per_Capita_USD"], df_clean["Infant_Mortality_Rate"])
plt.yticks([0, 50, 100], ["0%", "50%", "100%"])
plt.xlabel("GDP per capita")
plt.ylabel("Literacy Rate")
plt.title("Real GDP PPP billion USD vs Infant Mortality Rate", fontdict={"fontsize": 14}, weight="bold")
plt.grid(True, alpha=0.3)
plt.savefig("../../report/interim/stability_scatter.png", dpi=300, bbox_inches="tight")
plt.show()
```
